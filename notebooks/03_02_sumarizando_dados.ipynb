```python
import requests
import pandas as pd
```

```python
# Request URL
url_base = 'https://dadosabertos.camara.leg.br/api/v2/proposicoes?dataApresentacaoInicio=2025-07-01&dataApresentacaoFim=2025-10-01&ordem=ASC&ordenarPor=id'
```

```python
# Coleta dados
dados = requests.get(url_base).json()
```

```python
dados
```

```python
# Requisita os dados em formato json das primeiras 81 páginas

# Lista para armazenar todos os dados
lista_dados = []
offset = 1

# Loop para coletar todas as páginas
for offset in range(1,82):
    # URL com paginação
    url = f"{url_base}&pagina={offset}&itens=15"

    # Faz a requisição
    response = requests.get(url).json()
    records = response['dados']

    # # Adiciona os dados à lista
    lista_dados.extend(records)

    # # Próxima página
    offset += 1
```

```python
# Verifica a quantidade de registros coletados
len(lista_dados)
```

```python
# Cria dataframe pandas com os dados
df_proposicoes = pd.DataFrame.from_dict(data=lista_dados)
```
