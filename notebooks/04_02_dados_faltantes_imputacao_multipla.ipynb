```python
import pandas as pd
import kagglehub
from kagglehub import KaggleDatasetAdapter

from miceforest import ImputationKernel
```

```python
# Título do arquivo a ser descarregado
file_path = "Life Expectancy Data.csv"
```

```python
# Coleta de dados e armazenamento em dataframe Pandas
df = kagglehub.load_dataset(KaggleDatasetAdapter.PANDAS,"nailasrivastava/life-expectancy-analysis",file_path)
```

```python
df.head()
```

```python
# Imprimir informação sobre o dataframe
df.info()
```

```python
# Converte colunas para tipo string
df[['Country', 'Status']] = df[['Country', 'Status']].astype('category')
```

```python
df.columns
```

```python
# Gera dados para serem imputados
gera_dados = ImputationKernel(data = df, save_all_iterations_data = True, random_state = 2023)
```

```python
# Intera duas vezes no dataframe
gera_dados.mice(2)
```

```python
# Preenche os dados faltantes no dataframe
imputacao_multipla = gera_dados.complete_data()
```
